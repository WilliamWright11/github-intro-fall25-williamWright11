Question 23: Which variable pair has the highest pearson correlation in the whole iris dataset?

sepal_length, sepal_width
petal_length, sepal_length
sepal_length, petal_width
petal_length, petal_width

```python
# Calculate the correlation matrix
correlation_matrix = iris.corr(numeric_only=True)

# Get the correlation values for the specified pairs
sepal_length_sepal_width_corr = correlation_matrix.loc['sepal_length', 'sepal_width']
petal_length_sepal_length_corr = correlation_matrix.loc['petal_length', 'sepal_length']
sepal_length_petal_width_corr = correlation_matrix.loc['sepal_length', 'petal_width']
petal_length_petal_width_corr = correlation_matrix.loc['petal_length', 'petal_width']

# Store the correlations in a dictionary
correlations = {
    'sepal_length, sepal_width': sepal_length_sepal_width_corr,
    'petal_length, sepal_length': petal_length_sepal_length_corr,
    'sepal_length, petal_width': sepal_length_petal_width_corr,
    'petal_length, petal_width': petal_length_petal_width_corr
}

# Find the pair with the highest absolute correlation
highest_corr_pair = max(correlations, key=lambda item: abs(correlations[item]))
highest_corr_value = correlations[highest_corr_pair]

print(f"The variable pair with the highest Pearson correlation is '{highest_corr_pair}' with a correlation of {highest_corr_value:.3f}")
```

Quesiton 24:
Which of the following are true for whole iris dataset?

The mean value of petal_length of setosa is 1.46 (rounded)
setosa has the largest average sepal_width
The standard deviation of sepal_length of setosa is 0.35 (rounded)
virginica has largest sepal_length

```python
# Calculate mean petal_length for setosa
setosa_petal_length_mean = iris[iris['species'] == 'setosa']['petal_length'].mean()
print(f"Mean petal_length for setosa: {setosa_petal_length_mean:.2f}")

# Calculate average sepal_width for each species and find the max
avg_sepal_width_by_species = iris.groupby('species')['sepal_width'].mean()
print(f"Average sepal_width by species:\n{avg_sepal_width_by_species}")
setosa_has_largest_sepal_width = avg_sepal_width_by_species.idxmax() == 'setosa'
print(f"Setosa has the largest average sepal_width: {setosa_has_largest_sepal_width}")


# Calculate standard deviation of sepal_length for setosa
setosa_sepal_length_std = iris[iris['species'] == 'setosa']['sepal_length'].std()
print(f"Standard deviation of sepal_length for setosa: {setosa_sepal_length_std:.2f}")

# Find the species with the largest sepal_length
largest_sepal_length_species = iris.loc[iris['sepal_length'].idxmax(), 'species']
print(f"Species with the largest sepal_length: {largest_sepal_length_species}")
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

Question 25: What is the number of passengers in the 2nd class (pclass of 2) who survived (survived equal to 1) in the titanic dataset?

```python
# Filter for passengers in 2nd class who survived
second_class_survived = titanic[(titanic['pclass'] == 2) & (titanic['survived'] == 1)]

# Count the number of observations
num_second_class_survived = len(second_class_survived)

print(f"The number of passengers in the 2nd class who survived is: {num_second_class_survived}")
```

Question 26: What is the age range in the 1st class who did not survive (survived equal to 0)?
16-70
1-74
2-71
4-80

```python
# Filter for passengers in 1st class who did not survive
first_class_not_survived = titanic[(titanic['pclass'] == 1) & (titanic['survived'] == 0)]

# Get the age range
min_age = first_class_not_survived['age'].min()
max_age = first_class_not_survived['age'].max()

print(f"The age range for 1st class passengers who did not survive is: {min_age}-{max_age}")
```

Question 27: What is the percentage of female passengers who survived, rounded to the nearest whole percent?

```python
# Filter for female passengers
female_passengers = titanic[titanic['sex'] == 'female']

# Filter for female passengers who survived
female_survived = female_passengers[female_passengers['survived'] == 1]

# Calculate the percentage of female survivors
percentage_female_survived = (len(female_survived) / len(female_passengers)) * 100

# Round to the nearest whole percent
rounded_percentage = round(percentage_female_survived)

print(f"The percentage of female passengers who survived is: {rounded_percentage}%")
```

Question 29: Which of the following are true for titanic dataset?
Survival rate is highest in 1st class passengers
Survival rate is highest for passengers who embarked at S = Southampton
Survival rate is highest for passengers who emabarked at C = Cherbourg
Survival has positive correlation with passenger class
Survival has negative correlation with fare

```python
# Calculate survival rate by passenger class
survival_rate_by_class = titanic.groupby('pclass')['survived'].mean()
print(f"Survival rate by passenger class:\n{survival_rate_by_class}")

# Calculate survival rate by embarkation point
survival_rate_by_embarked = titanic.groupby('embarked')['survived'].mean()
print(f"Survival rate by embarkation point:\n{survival_rate_by_embarked}")

# Calculate correlation between survival and passenger class
survival_pclass_corr = titanic['survived'].corr(titanic['pclass'])
print(f"Correlation between survival and passenger class: {survival_pclass_corr:.3f}")

# Calculate correlation between survival and fare
survival_fare_corr = titanic['survived'].corr(titanic['fare'])
print(f"Correlation between survival and fare: {survival_fare_corr:.3f}")
```
